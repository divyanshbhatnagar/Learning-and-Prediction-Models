```r
library(clusterSim)
```

```r
data_new <- data.Normalization (data,type="n4",normalization="column")
```

```r
data_new
```

## With 5 clusters

```r
set.seed(76964057) 
k <-kmeans(data_new, centers=5)
k$centers
```

```r
table(k$cluster)
```

We conclude that in the 1st Cluster the people use internet, own a smartphone, have used dating app, play games and are not a student

## With 3 clusters

```r
k <-kmeans(data_new, centers=3)
```

```r
k$centers
```

```r
table(k$cluster)
```

We conclude that in Cluster 1 people use internet, have smartphone, do not use dating app, play games and are not a student
